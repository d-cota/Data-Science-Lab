```python
import csv

wrapper = []

with open('online_retail.csv') as f:
    reader = csv.reader(f)
    next(reader)
    [wrapper.append(item) for item in reader if item[0].isnumeric()]

len(wrapper)
```

```python
invoice_dict = {}
products = []

for row in wrapper:
    invoice = row[0]
    if invoice not in invoice_dict:
        invoice_dict[invoice] =  [row[2]]
    else:
        invoice_dict[invoice].append(row[2])
    if row[2] not in products: products.append(row[2])

products.sort()
len(invoice_dict)
```

```python
pa_matrix = []
for invoice in invoice_dict:
    temp = []
    for p in products:
        if p in invoice_dict[invoice]: temp.append(1)
        else: temp.append(0)
    pa_matrix.append(temp)

pa_matrix[0]
```

```python
import pandas as pd

df = pd.DataFrame(data=pa_matrix ,columns=products)
df

```

```python
from mlxtend import frequent_patterns

fi = frequent_patterns.fpgrowth(df, 0.02)

print(len(fi))
print(fi.to_string())
```

```python
ar = frequent_patterns.association_rules(fi, min_threshold=0.8)
ar
```

```python
fi_apriori = frequent_patterns.apriori(df, min_support=0.02)
print(len(fi_apriori))
print(fi_apriori.to_string())
```

```python
import timeit

timeit.timeit(lambda: frequent_patterns.apriori(df, 0.02), number=1), timeit.timeit(lambda: frequent_patterns.fpgrowth(df, 0.02), number=1)
```
